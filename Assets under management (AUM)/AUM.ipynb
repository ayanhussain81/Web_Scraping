# Open End Schemes

```python
from  bs4 import BeautifulSoup 
import requests
from datetime import date
import pandas as pd
import numpy as np
import pyodbc
```

```python
month="4"
year="2022"
end_date = {'1': '31', '2': '28', '3': '31','4': '30','5': '31','6': '30','7': '31','8': '31','9': '30','10': '31','11': '30','12': '31'}

try:
    url="https://mufap.com.pk/aum_report.php?tab=01&amc=&fname=&cat=&mnt="+str(month)+"&yrs="+str(year)+"&submitted=Show+Report";
    print(url)
    page = requests.get(url)
except:
    print("Wrong url")
```

```python
day = end_date[str(month)]
```

```python
ValidityDate = '%s-%s-%s' % (year,month, day)
```

```python
ValidityDate
```

```python
Soup = BeautifulSoup(page.content,'html.parser')
```

```python
data1 = []
```

```python
table = Soup.find('table',class_="mydata")
try:
    for row in table.find_all('tr',class_=['tab-data border','tab-data border report-row-back']):
            cols = row.find_all('td')
            data1.append((cols[0].text.strip(), cols[1].text.strip(), cols[2].text.strip(), cols[3].text.strip()))
except:
    pass
```

```python
data1
```

```python
scrap_data = np.asarray(data1)
len(scrap_data)
```

```python
df1 = pd.DataFrame(scrap_data)
```

```python
df1.head(10)
```

```python
df.shape
```

```python
i = df[df['Category'] == 'Fund of Funds' ].index
j = df[df['Category'] == 'Shariah Compliant Fund of Funds' ].index
k = df[df['Category'] == 'Shariah Compliant Fund of Funds - CPPI' ].index
l = df[df['Category'] == 'Fund of Funds - CPPI' ].index
m = df[df['Category'] == 'Shariah Compliant Fund of Funds - Income' ].index
```

```python
df.drop(i, inplace = True)
df.drop(j, inplace = True)
df.drop(k, inplace = True)
df.drop(l, inplace = True)
df.drop(m, inplace = True)
```

```python
df.head()
```

```python
for x in df['AUM']:  
       df['AUM']=df['AUM'].replace("Not Published","0")
```

```python
df
```

```python
df['MonthYear']=ValidityDate
```

```python
df
```

```python
df.drop("InceptionDate", axis=1, inplace=True)
```

```python
df
```

```python
df['AUM']=df['AUM'].str.replace(',','')
```

```python
df['AUM'] = df['AUM'].astype(float)
```

```python
df['AUM'] = 1000 * df['AUM']
```

```python
df
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace = True)
df= df.fillna(0)
```
