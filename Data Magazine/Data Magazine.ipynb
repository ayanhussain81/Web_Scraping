```python
import requests
from bs4 import BeautifulSoup
from csv import writer
from lxml import html
```

```python
url = "https://datamagazine.co.uk/15-best-pakistan-big-data-startups-companies/"
page = requests.get(url)
tree = html.fromstring(page.content)
```

```python
Soup = BeautifulSoup(page.content, 'html.parser')
lists = Soup.find('div', class_='g1-content-narrow').find_all('div',class_='wp-block-cover')
print(lists)
```

```python
with open('Data_csv', 'w', encoding= 'utf8' ,newline='') as f:
    thewriter = writer(f)
    header = ["Name", "Crunchbase", "Website", "Twitter","Facebook", "Linkedin","Desciption"]
    thewriter.writerow(header)
    info =[]
    i = 2
    for list in lists:
        list1=[]
        name = list.find('span').text
        list1.append(name)
        pp = list.find('p')
        links = []
        for link in pp.find_all('a'):
            if 'https' not in link['href']:
                list1.append("None")
            else:
                list1.append(link['href'])
        i+=1
        des = tree.xpath('//*[@id="post-12152"' + ']/div[1]/div[2]/div[' + format(i) + ']/div/p[2]/text()')
        list1.append(des)
        info.append(list1)
        thewriter.writerow(info)
    print("Data has been extened")
```
