```python
def employee(emp):
    new = emp.split('on')
    data  = new[0]
    return data



with open('DataCompanies.csv', 'a', encoding= 'utf8' ,newline='') as f:
    thewriter = writer(f)
    header= ["link","Overview","Website","Company_Size","Employees on Linkedin","Indutry","Founded","Location","Specialties"]

    thewriter.writerow(header)         
    with open('linksss1.csv', 'r') as file:
        reader = csv.reader(file)
        for count, row in enumerate(reader):
            for i in row:
                web,year,location,overview,company_size='','','','',''
                indutry,Specialties,employees='','',''

                driver.get(i)
                time.sleep(5)
                src = driver.page_source

                soup = BeautifulSoup(src, 'lxml')
                try:
                # intro = soup.find('section', {'class': 'artdeco-card p5 artdeco-card mb4'})
                    overview = soup.find('p', {'class': 'white-space-pre-wrap'}).text.strip()
                except: 
                    pass
                try:
                    company_size = soup.find('dd',class_='t-black--light text-body-small mb1').text.strip()
                except: 
                    pass
                try:
                    employ = soup.find('dd',class_='t-black--light mb4 text-body-small').text.strip()
                    employees = employee(employ)
                except: 
                    pass
                try:
                    texts = soup.find_all('dd',class_='mb4 t-black--light text-body-small')
                    Specialties = texts[-1].text.strip()

                    for t in texts:
                        a = t.text.strip()
                        if 'http' in a :
                            web = a
                        if len(str(a)) == 4:
                            year = a
                except:
                    pass
                try:
                    intro = soup.find_all('div',class_='org-top-card-summary-info-list__info-item')
                    indutry = intro[0].text.strip()
                    if 'followers' not in intro[1].text:
                        location = intro[1].text.strip()
                except:
                    pass
                    
                info = [i,overview,web,company_size,employees,indutry,year,location,Specialties]
                thewriter.writerow(info)
                print(i,count)
```
