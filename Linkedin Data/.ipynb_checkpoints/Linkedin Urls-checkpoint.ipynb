```python
import requests
from bs4 import BeautifulSoup
from googlesearch import search

# Define the search query and company name
query = "linkedin company name"
company_name = "Saudi Basic Industries (SABIC)"

# Search for the query on Google
for url in search(query + company_name, stop=5):
    # Check if the URL is a LinkedIn page
    if "linkedin.com/company/" in url:
        # Get the HTML content of the page
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        # Find the company name and check if it matches the input name
        name = soup.find("h1", class_="org-top-card-summary__title").text.strip()
        if company_name.lower() in name.lower():
            # Print the LinkedIn page URL
            print("LinkedIn URL:", url)
            break
```

```python
from googlesearch import search
import re

# Enter the company name you want to search for
query = "Saudi Basic Industries"

# Perform the Google search and get the first result
search_results = search(query, num_results=1)

# Extract the LinkedIn link from the search result using regular expressions
for result in search_results:
    match = re.search("(?P<url>https?://[^\s]+)", result)
    if match is not None:
        url = match.group("url")
        if "linkedin.com/company/" in url:
            print(url)
            break
```

```python
from googlesearch import search
from bs4 import BeautifulSoup
import requests

# Query to search on Google
query = "Saudi Basic Industries Linkedin"

# Search on Google and get the first result
url = next(search(query))

# Get the HTML content of the webpage
response = requests.get(url)
html_content = response.text

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find the LinkedIn link
linkedin_link = soup.find('a', {'class': 'linkedin'})['href']
print(linkedin_link)
```

```python
with open('Extractedlinkedinlinks.csv', 'a', encoding= 'utf8' ,newline='') as f:
    thewriter = writer(f)
    with open('linkedinlinks.csv', 'r') as file:
        reader = csv.reader(file)
        for count, row in enumerate(reader):
            for i in row:
                query = i
                # Search on Google and get the first result
                urls = search(query,  num=1)

                # Get the first URL from the search results
                url = next(urls)
                info = [url,i]
                # Print the first link
                thewriter.writerow(info)
                print(url,count)
```
