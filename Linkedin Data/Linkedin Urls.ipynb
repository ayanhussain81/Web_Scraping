```python
import csv
from _csv import writer
import json
from googlesearch import search
```

```python
with open('Extractedlinkedinlinks.csv', 'a', encoding= 'utf8' ,newline='') as f:
    thewriter = writer(f)
    with open('linkedinlinks.csv', 'r') as file:
        reader = csv.reader(file)
        for count, row in enumerate(reader):
            for i in row:
                query = i
                # Search on Google and get the first result
                urls = search(query,  num=1)

                # Get the first URL from the search results
                url = next(urls)
                info = [url,i]
                # Print the first link
                thewriter.writerow(info)
                print(url,count)
```
